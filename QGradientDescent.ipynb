```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def LossFunction(x):
    return 4 * x ** 3 + 6  * x ** 2 - 3 * x - 3

def DifferentialLossFunction(x):
    return 12  * x ** 2 + 12 * x - 3

def GradientDescent(initX, rate, maxIters):
    iters = 0
    while iters < maxIters:
        dx = DifferentialLossFunction(initX)
        initX -= dx * rate
        iters+=1
    return initX
```

```python
x = np.linspace(-5, 25, 100000)
y = LossFunction(x)

plt.figure(figsize=(6, 4), dpi=100)
plt.plot(x, y)
plt.title('Loss Fuction')
```

```python
initX = np.random.randint(-10, 10)
rate = 0.00001 # Learning rate
maxIters = 10000

curX = round(GradientDescent(initX, rate, maxIters), 5)
print(curX)
```

```python
plt.figure(figsize=(6, 4), dpi=100)
plt.plot(x, y)
plt.scatter(curX, LossFunction(curX), color='Red' )
```
